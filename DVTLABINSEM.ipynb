```python
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sample tweet data
sample_tweets = [
    "I support the #BlackLivesMatter movement.",
    "Equality and justice for all. #BLM",
    "Discussing systemic racism in our society.",
    "Racial injustice must end.",
    "Solidarity with the fight against police brutality.",
    "We stand together for human rights.",
    "Promoting social change.",
    "Injustice anywhere is a threat to justice everywhere. #BlackLivesMatter"
]

# Data preprocessing
def preprocess_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+|\@\w+|\#\w+", '', text)
    text = re.sub(r"[^a-zA-Z\s]", '', text)
    text = text.lower()
    return text

cleaned_tweets = [preprocess_text(tweet) for tweet in sample_tweets]

# Frequency analysis
word_counts = Counter()
for tweet in cleaned_tweets:
    words = tweet.split()
    word_counts.update(words)

# Visualize word frequencies
common_words = word_counts.most_common(10)
word_freq_dict = dict(common_words)

# Create a bar chart
plt.bar(word_freq_dict.keys(), word_freq_dict.values())
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Words in Sample Tweets')
plt.xticks(rotation=45)
plt.show()

# Create a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq_dict)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Sample Tweets')
plt.show()
```
